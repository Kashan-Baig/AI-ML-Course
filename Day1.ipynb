# **Sicket-Learn Basics**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = sns.load_dataset('tips')
df.head()
```

## **Introduction To Linear Regression**

### **Analysing with Graph**

```python
#regression model graph
sns.lmplot(x='total_bill', y='tip', data=df)
```

### **Taking 1 Variable as independent**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[['total_bill']]
y = df['tip']
model.fit(X, y)
```

```python
# prdicting the tip for a total bill of 20
print("Total bill = ",model.predict([[20]]))
```

### **Taking 2 Variable as independent**

```python
model2 = LinearRegression()
X2 = df[['total_bill', 'size']]
y2 = df['tip']
model2.fit(X2, y2)
# prdicting the tip for a total bill of 20 and size of 3
print("Total bill = ",model2.predict([[20, 3]]))
```

### **Taking 3 Variable (Categorical)** 

```python
df['smoker_num'] = df['smoker'].map({'Yes': 1, 'No': 0})
model3 = LinearRegression()
X3 = df[['total_bill', 'size','smoker_num']]
y3 = df['tip']
model3.fit(X3, y2)
           
# prdicting the tip for a total bill of 20 and size of 3
print("Total bill = ",model3.predict([[20, 3, 0]]))
```

## **Introduction to Logistic Regression**

```python
data = sns.load_dataset('tips')
x = data[['total_bill','size','tip']]
y = data['smoker']

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x,y)
# prdicting the smoker status for a total bill of 20 and size of 3
print("Smoker status = ",lr.predict([[20, 3, 5]]))
```

## **Naive bayes**

```python
# **Introduction to Naive bayes*
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
X = df[['total_bill', 'size','tip']]
y = df['smoker']
gnb.fit(X, y)

# prdicting the smoker status for a total bill of 20 and size of 3
print("Smoker status = ", gnb.predict([[20, 3, 5]]))
```

## **Decision Tree**

```python
from sklearn.tree import  DecisionTreeRegressor
dt = DecisionTreeRegressor()
X = df[['total_bill', 'size']]
y = df['tip']
dt.fit(X, y)
# prdicting the tip for a total bill of 20 and size of 3
print("Total bill = ", dt.predict([[20, 3]]))
```
